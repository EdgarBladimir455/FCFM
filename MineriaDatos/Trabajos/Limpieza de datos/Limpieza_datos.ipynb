Eliminar los registros con datos nulos

```python
vgSalesOriginal = vgSalesOriginal.dropna()
vgSalesOriginal.isnull().sum()
```

## Cambiar nombres de columnas

```python
vgSalesOriginal = vgSalesOriginal.rename(columns = {'Name':'NombreJuego'})
vgSalesOriginal = vgSalesOriginal.rename(columns = {'Platform':'Consola'})
vgSalesOriginal = vgSalesOriginal.rename(columns = {'Year':'AnnoPublicacion'})
vgSalesOriginal = vgSalesOriginal.rename(columns = {'Genre':'Genero'})
vgSalesOriginal
```

## Agrupacion de datos por genero de juego

Generos existentes en la base

```python
list(vgSalesOriginal.Genero.unique())
```

Agrupacion de registros por tipo de genero **Plataforma**

```python
vgSalesGeneroPlataforma = vgSalesOriginal[vgSalesOriginal['Genero'] == 'Platform']
vgSalesGeneroPlataforma
```

Agrupacion de registros por consola **SNES**

```python
vgSalesConsolaSNES = vgSalesOriginal[vgSalesOriginal['Consola'] == 'SNES']
vgSalesConsolaSNES
```

## Generar archivos con los datos agrupados

```python
vgSalesGeneroPlataforma.to_csv('vgSalesGeneroPlataforma.csv', index=False)
vgSalesConsolaSNES.to_csv('vgSalesConsolaSNES.csv', index=False)
```

* Investigar los siguientes comandos: ?df.dropna() ?df.astype

**dropna():** remueve valores nulos o vacios
**astype:** sirve para castear el tipo de dato para el dataframe o cada columna

**df=df.replace('\*','',regex=True).astype(float)**: remplaza todos los asteriscos por espacios en blanco, despues se castea el tipo de dato a float lo que causa un error ya que no se puede castear de espacio en blanco a un float.

**df.dropna(thresh=int(df.shape[0] * .9), axis=1)**: elimina los registros que tengan valores nulos o vacios, al usar thresh se especifica la cantidad de valores NO-nulos minimos que debe tener el registro para mantenerlo, y al usar axis=1 se especifica que la cantidad de NO-nulos debe ser pero a nivel columnas, es decir la candidad de registros por .9 es la cantidad minima de registros NO-nulos que debe tener la columna para mantenerla, en caso contrario se elimina la columna.
