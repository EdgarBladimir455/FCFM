## Librerias y carga de la base de datos

```python
import pandas as pandas
import seaborn as sns
import matplotlib.pyplot as plt

vgSalesOriginal = pandas.read_csv("vgsales.csv",encoding='utf8')
```

## Tipos de datos de las columnas

```python
vgSalesOriginal.dtypes
```

## Eliminar los registros nulos

```python
vgSalesOriginal = vgSalesOriginal.dropna()
vgSalesOriginal.isnull().sum()
```

## Descripcion estadistica basica

Se establece la columna fecha como enteros

```python
vgSalesOriginal.Year.astype("Int32")
```

Se eliminan los datos no decimales, se agrupa por genero y se saca la suma total de ventas segun los generos

```python
vgSalesGeneroSumatoriaVentas = vgSalesOriginal.drop(['Rank', 'Name', 'Year', 'Platform', 'Publisher'],axis = 1 )
vgSalesGeneroSumatoriaVentas = vgSalesGeneroSumatoriaVentas.groupby(['Genre'], as_index=False).sum()
vgSalesGeneroSumatoriaVentas
```

Se eliminan los datos no decimales, se agrupa por plataforma y se saca un promedio de ventas.

```python
vgSalesPlataformaSumatoriaVentas = vgSalesOriginal.drop(['Rank', 'Name', 'Year', 'Genre', 'Publisher'],axis = 1 )
vgSalesPlataformaSumatoriaVentas = vgSalesPlataformaSumatoriaVentas.groupby(['Platform'], as_index=False).sum()
vgSalesPlataformaSumatoriaVentas
```

Se eliminan los datos no decimales, se agrupa por desarrollador/distribuidor y se sacan los valores maximos de ventas que ha ganado por un juego.

```python
vgSalesPublisherVentasMaximas = vgSalesOriginal.drop(['Rank', 'Name', 'Year', 'Platform', 'Genre'],axis = 1 )
vgSalesPublisherVentasMaximas = vgSalesPublisherVentasMaximas.groupby(['Publisher'], as_index=False).max()
vgSalesPublisherVentasMaximas
```

Se eliminan los datos no decimales, se agrupa por fecha y se sacan la suma total de ventas que hubo en ese año

```python
vgSalesYearSumatoriaVentas = vgSalesOriginal.drop(['Rank', 'Name', 'Genre', 'Platform', 'Publisher'],axis = 1 )
vgSalesYearSumatoriaVentas = vgSalesYearSumatoriaVentas.groupby(['Year'], as_index=False).sum()
vgSalesYearSumatoriaVentas
```

## Grafica de cantidad de ventas por genero de juego

```python
vgSalesOriginal['Genre'].value_counts().plot.bar()
```

* El genero de juego que mas se vende es el de Acción, y el menos vendido es el de Puzzle

## Grafica de cantidad de ventas a traves de los años

```python
vgSalesYearSumatoriaVentas.plot(x='Year')
```

* Des los años 2005 a 2015 fueron los años con mas ventas, de 2015 en adelante hubo una baja en las ventas (puede ser por falta de informacion)

## Grafica que muestra la venta de videojuegos entre todas los desarrolladores/distribuidores en america comparada con las ventas globales

```python
sns.jointplot(x="Global_Sales", y="NA_Sales", data=vgSalesPublisherVentasMaximas,
                  kind="reg", truncate=False,
                  xlim=(0, 60), ylim=(0, 12),
                  color="m", height=7)
```

* La mayoria de ventas de un juego entre el conjunto de desarrolladores/distribuidores de las ventas globales son en america, es decir que de las ventas globales de un juegos de una desarrolladora, la mayor parte de sus ventas es en america

```python
f, ax = plt.subplots(figsize=(20, 15))

sns.set_color_codes("pastel")
sns.barplot(x="Global_Sales", y="Platform", data=vgSalesPlataformaSumatoriaVentas,
            label="Ventas Globales", color="b")

sns.set_color_codes("muted")
sns.barplot(x="NA_Sales", y="Platform", data=vgSalesPlataformaSumatoriaVentas,
            label="Ventas en America", color="b")

sns.despine(left=True, bottom=True)
```

* La ps2 es la que mas inversion tuvo respecto a desarrollo de juegos y la mitad de sus ventas pertencen a la region de america y la otra mitad distribuida entre las demas regiones

**SCATTER PLOT**
* Muestra el valor de 2 conjuntos de datos en 2 dimensiones
* Cada punto representa una observación 
* Es posible asignar otra variable al tamaño de cada punto, lo que hace un bubble plot
* Si tiene muchos puntos y tiene problemas con la sobreimpresión, considere usar un 2D density plot.

**BOXPLOT**
* Ofrece un buen resumen de una o varias variables numéricas
* La línea que divide la caja en 2 partes representa la mediana de los datos
* El final del cuadro muestra los cuartiles superior e inferior
* Las líneas extremas muestran el valor más alto y más bajo excluyendo los valores atípicos
